This notebook is to test cancer_models.py and to show an example of how to use MLE on the patient's data to get the models'parameters.

```python
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm 

from scipy.optimize import minimize

%pip install pyswarms --quiet
from pyswarms.single.global_best import GlobalBestPSO
```

```python
from cancer_models import *
from libraries.dataset import *
```

Test all the models on one patient

```python
patient = "C233"
y_obs = np.array(mtm_dataset[patient])
times = np.array(time_dataset[patient])
drug_times = np.array(time_drug_dataset[patient]) - times[0]
times = times - times[0]
```

Test PSO+LBFGS

```python
all_models =  ["exp","log","bert","gomp","exp_stairs","sls","sls*","slsv2", "fs_exp"] #bert_stairs out since it needs SLSQP

for model_name in all_models:
    print(f"model: {model_name}")

    bounds_for_model = np.array(get_bounds(model_name,y_obs[0]))
    bounds_PSO_format = (bounds_for_model[:,0],bounds_for_model[:,1])
    cost = []
    pos = []

    #PSO:
    options = {'c1': 0.9, 'c2': 0.9, 'w': 1.1} #PSO parameters
    PSO = GlobalBestPSO(n_particles=80, dimensions=len(bounds_for_model), options=options, bounds=bounds_PSO_format)
    cost_PSO, pos_PSO = PSO.optimize(super_Q_wrapper, 300, model_name=model_name, times=times, y_obs=y_obs, drug_times=drug_times)

    cost_mask = np.array(PSO.swarm.current_cost).argsort()
    x = PSO.swarm.position[cost_mask][:20,:] #take the 20 best positions

    #LBFGS:
    for i in tqdm(range(len(x))):
        result = minimize(objective_wrapper, x[i], args = (model_name, times, y_obs, drug_times,), bounds = bounds_for_model, method = 'l-bfgs-b')
        cost.append(result.fun)
        pos.append(result.x)
    
    #best position:
    cost = np.array(cost)
    best_l = np.nanmin(cost)
    print(f"best cost:{best_l}")

    min_cost = np.where(cost == np.nanmin(cost))[0][0]
    best = pos[min_cost]

    #plot
    sigma = best[0]
    y_pred = model_wrapper(model_name, best[1:], y_obs, times, drug_times)
    plt.plot(np.linspace(0,times[-1],len(y_pred)),np.log(y_pred+eps), label=f"{model_name}")
    plt.fill_between(np.linspace(0,times[-1],len(y_pred)),np.log(y_pred+eps)-sigma, np.log(y_pred+eps)+sigma, color='gray', alpha=0.3)
    plt.scatter(times,np.log(y_obs+eps), label="observed data")
    plt.legend()
    plt.show()
```

Test Bert stairs

```python
model_name = 'bert_stairs'

bounds_for_model = np.array(get_bounds(model_name,y_obs[0]))
bounds_PSO_format = (bounds_for_model[:,0],bounds_for_model[:,1])
cost = []
pos = []

#generate grid:
n_grid = 3
starting_pos = np.array([])
for j in range(len(bounds_for_model)):
    starting_pos = np.append(starting_pos, np.linspace(bounds_for_model[j][0], bounds_for_model[j][1], n_grid+2)[1:-1])
starting_pos = np.reshape(starting_pos, (len(bounds_for_model), -1))
grid = np.meshgrid(*[starting_pos[z] for z in range(len(bounds_for_model))])
x = np.array(grid).T.reshape(-1,len(bounds_for_model))

#define the constraint (carrying capacity above 1.1*y_{obs}(0))
con = {'type': 'ineq', 'fun': lambda y: np.array((y[2]/y[3])**3 - 1.1*y[1])}

# randomly shuffle the grid points and then choose the first 70 ones
index = np.random.choice(x.shape[0], 70, replace=False)

#run SLSQP on those 70 points
for i in tqdm(range(70)):
    if (x[index[i]][2]/x[index[i]][3])**3 < 1.1*x[index[i]][1]:
        x[index[i]][3] = np.random.rand(1)[0] * x[index[i]][2]/(x[index[i]][1]**(1/3))
    res = minimize(objective_wrapper, x[index[i]], args = ('bert_stairs', times, y_obs, drug_times,), method="SLSQP", jac="2-point", bounds=bounds_for_model, constraints = con)
    cost.append(res.fun)
    pos.append(res.x)

#best position:
cost = np.array(cost)
best_l = np.nanmin(cost)
print(f"best cost:{best_l}")

min_cost = np.where(cost == np.nanmin(cost))[0][0]
best = pos[min_cost]

#plot
sigma = best[0]
y_pred = model_wrapper(model_name, best[1:], y_obs, times, drug_times)
plt.plot(np.linspace(0,times[-1],len(y_pred)),np.log(y_pred+eps), label=f"{model_name}")
plt.fill_between(np.linspace(0,times[-1],len(y_pred)),np.log(y_pred+eps)-sigma, np.log(y_pred+eps)+sigma, color='gray', alpha=0.3)
plt.scatter(times,np.log(y_obs+eps), label="observed data")
plt.legend()
plt.show()
```
